```python
cars= pd.read_csv("cars.csv")
cars.drop("Unnamed: 0",axis=1,inplace=True)
cars
```

```python
car = list(cars["Description"])
```

```python
#CArwale

driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()  #maximizing windows


#empty box
model=[]
price=[]
year=[]
km=[]
ft=[]
tm=[]
place=[]
color=[]
nof=[]
insu=[]
ltt=[]

#extracting data
for i in car:
    driver.get(i)
    time.sleep(3)
    
#Model
    try:
        c=driver.find_element_by_xpath("//*[@id='content-title']")
        model.append(c.text)
    except NoSuchElementException:
        model.append("--")
#price
    try:
        c=driver.find_element_by_xpath("//*[@id='overview']/div/ul/li[1]/div[2]")
        price.append(c.text)
    except NoSuchElementException:
        price.append("--")

#year
    try:
        z=driver.find_element_by_xpath("//*[@id='overview']/div/ul/li[2]/div[2]")
        year.append(z.text)
    except NoSuchElementException:
        year.append("--")

#Kilometer
    try:
        d=driver.find_element_by_xpath('//*[@id="overview"]/div/ul/li[3]/div[2]')
        km.append(d.text)
    except NoSuchElementException:
        km.append("--")
#feaul type

    try:
        c=driver.find_element_by_xpath("//*[@id='overview']/div/ul/li[4]/div[2]")
        ft.append(c.text)
    except NoSuchElementException:
        ft.append("--")

#transmision
    try:
        c=driver.find_element_by_xpath("//*[@id='overview']/div/ul/li[5]/div[2]")
        tm.append(c.text)
    except NoSuchElementException:
        tm.append("--")
        
#place
    try:
        c=driver.find_element_by_xpath("//*[@id='available-text']")
        place.append(c.text)
    except NoSuchElementException:
        place.append("--")
        
#color
    try:
        c=driver.find_element_by_xpath("//*[@id='overview']/div/ul/li[7]/div[2]")
        color.append(c.text)
    except NoSuchElementException:
        color.append("--")
        
#number of owner
    try:
        c=driver.find_element_by_xpath("//*[@id='overview']/div/ul/li[9]/div[2]")
        nof.append(c.text)
    except NoSuchElementException:
        nof.append("--")
        
#insurance
    try:
        c=driver.find_element_by_xpath("//*[@id='overview']/div/ul/li[11]/div[2]")
        insu.append(c.text)
    except NoSuchElementException:
        insu.append("--")
    
#Life time tax
    try:
        c=driver.find_element_by_xpath("//*[@id='overview']/div/ul/li[12]/div[2]")
        ltt.append(c.text)
    except NoSuchElementException:
        ltt.append("--")
    
time.sleep(2)
#creating dataframe
df=pd.DataFrame({})
df["Model"]=model
df["Price"]=price
df["Year"]=year
df["kilo meter"]=km
df["Fuel Type"]=ft
df["Transmission"]=tm
df["Place"]=place
df["Color"]=color
df["NUmer of Owner"]=nof
df["Insurance"]=insu
df["Life Time Tax "]=ltt



df
```
